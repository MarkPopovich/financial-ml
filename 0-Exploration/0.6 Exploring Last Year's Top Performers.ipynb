````python
stock_stats = {}
````

````python
stocks_list = list(df['Symbol'])
````

````python
length = len(stocks_list)
````

````python
for i, stock in enumerate(stocks_list):
    df = extract_stock(stock, data_dir=data_dir, return_df=True, suffix=suffix)
    mean_return = df['pct_change'].mean()
    std_return = df['pct_change'].std()
    
    stock_stats[stock] = {'mean' : mean_return,
                          'std' : std_return}
    del(df)
    stocks_list.remove(stock)
    
    display.clear_output()
    pct = i/length *100
    print('{:.2f}%'.format(pct))
    time.sleep(.5)
````

````python
len(stocks_list)
````

````python
results = pd.DataFrame(stock_stats).T.sort_values(by='mean', ascending=False)
````

````python
results
````

````python
results['mean'] = results['mean'] * 100
results['std'] = results['std'] * 100
````

````python
results['alpha'] = results['mean'] / results['std']
````

````python
results = results.sort_values(by='alpha', ascending=False)
````

````python
pd.set_option('display.max_rows', 100)
````
