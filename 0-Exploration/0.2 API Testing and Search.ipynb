# API Testing and Search

In this notebook I will iteratively search for all instruments and add them to a list to be saved. 

```python
import requests
api_key = ''
projection = 'symbol-search' 
sym = 'X'

endpoint = r'https://api.tdameritrade.com/v1/instruments'

payload = {'apikey': api_key,
          'projection': projection,
          'symbol': sym}

res = requests.get(url = endpoint, params = payload)
res.json()
```

```python
res.json()[sym]
```

```python
from time import sleep

def test_sym(symbol, endpoint = endpoint, payload = payload):
    payload = {'apikey': api_key,
              'projection': projection,
              'symbol': symbol}
    
    res = requests.get(url = endpoint, params = payload)
    
    try:
        return res.json()[symbol]
    except KeyError:
        if 'error' in res.json().keys():
            if 'transactions per seconds' in res.json()['error']: 
                print('Timeout')
                sleep(1)
                test_sym(symbol)
        else:
            print("Error at Symbol: ", symbol)
            print(res.json())
            return res.json()
```

```python
test_sym('X')
```

```python
from itertools import product
from time import time

letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

x = 1
prods = []
start = time()
while x < 4:
    print(time() - start)
    prods.extend([''.join(prod) for prod in product(letters, repeat = x)])
    x+=1
prods
```

```python
prods[-1]
```

```python
responses = {}

for sym in prods:
    responses[sym] = test_sym(sym)
```
