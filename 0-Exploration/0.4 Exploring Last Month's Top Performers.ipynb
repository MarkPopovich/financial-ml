```python
stocks_list = list(df['Symbol'])
stock_stats = {}
length = len(stocks_list)
i = 0
```

```python
length
```

```python
stocks_list
```

```python
for stock in stocks_list:

    stock_data = get_history(stock, periodType='month', frequencyType='daily', frequency=1, periods=2)
    if stock_data['empty'] == True:
        stocks_list.remove(stock)
        continue
    stock_df = format_date(stock_data)
    stock_df['SYMBOL'] = stock
    stock_df = stock_df[::-1]
    stock_df['diff_close'] = stock_df['close'].diff(-1)
    stock_df['pct_close'] = stock_df['diff_close'] / stock_df['close'].shift(-1) * 100
    # calculate the average daily percent change and std dev.
    mean_return = stock_df['pct_close'].mean()
    std_return = stock_df['pct_close'].std()
    stock_df.to_pickle('./data/screens/volume600k/{}081820.pickle'.format(stock))
    # save the stock df as a pickle
    # save the mean % change and std to another dataframe for sorting 
    stock_stats[stock] = {'mean' : mean_return,
                          'std' : std_return}
    del(stock_df)
    stocks_list.remove(stock)
  
    display.clear_output()
    pct = i/length *100
    print('{:.2f}%'.format(pct))
    time.sleep(.5)
    i += 1
    
    
pd.DataFrame(stock_stats).T
```

```python
stocks_list
```

```python
results = pd.DataFrame(stock_stats).T.sort_values(by='mean', ascending=False)
```

```python
results
```

```python
results['alpha'] = results['mean'] / results['std']
```

```python
results = results.sort_values(by='alpha', ascending=False)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
results.dropna(inplace=True)
```

```python
c_res = results[results['alpha'] > 0.2]
```

```python
results.to_pickle('./data/screens/volume600k/results.pickle')
```
